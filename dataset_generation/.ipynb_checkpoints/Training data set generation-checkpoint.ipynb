# Training set generation
## C+ class

```python
%matplotlib inline
from matplotlib.pyplot import *
from matplotlib import pyplot as plt
import numpy as np
from importlib import reload

import vip_hci as vip
from hciplot import plot_frames, plot_cubes

from vip_hci.config import VLT_NACO
from vip_hci.fm import normalize_psf, cube_inject_companions
from vip_hci.psfsub import median_sub, pca, pca_annular, pca_annulus

from vip_hci.fits import open_fits, write_fits, info_fits
from vip_hci.metrics import significance, snr, snrmap
from vip_hci.var import fit_2dgaussian, frame_center

from vip_hci.preproc.cosmetics import cube_crop_frames, frame_crop
from vip_hci.preproc import frame_rotate

import training_set_generation_functions
reload(training_set_generation_functions)
from training_set_generation_functions import inject_random_fake_comp, make_mlar_plus, make_mlar_minus, get_fwhm, evaluate_snr, plot_patch, flux_interval, patch_rotation, patch_shift, patch_average
```

```python
adi=np.load('adi_seq.npy')
pa=np.load('pa.npy')
psf=np.load('psf.npy')
pxscale=np.load('pxscale.npy')

fwhm = get_fwhm(psf)
an_nbr=5
an_radius = an_nbr*fwhm
snr_interv=(1,3)
ncomp= 12
sample_nbr=100
flux_stat=50
```

```python
inj_adi=[]                 #keep adi cubes that have been injected with a satisfying S/N value
snr_values=[]
good_values=[]             # S/N, flux, injection position corrseponding to adi cubes in inj_adi
patches_plus=[]
for i in range(flux_stat):
    adi_with_fake_comp, inj_posxy = inject_random_fake_comp(adi, psf, pxscale, pa, fwhm, an_radius)
    snr_vale, f_source, sourcex, sourcey = evaluate_snr(adi_with_fake_comp, pa, fwhm, inj_posxy, an_radius)
    snr_values.append([snr_vale, f_source, sourcex, sourcey])
    if snr_vale>snr_interv[0] and snr_vale<snr_interv[1]:
        inj_adi.append(adi_with_fake_comp)
        good_values.append([snr_vale, f_source, sourcex, sourcey])
snr_values=np.array(snr_values)
```

Controlling S/N after random injections and evaluating mean flux interval to retreive the wanted S/N

```python
scatter(snr_values[:,1],snr_values[:,0])
flux_interv = flux_interval(snr_interv, snr_values)
```

```python
for i in range(len(inj_adi)):
    mlar=make_mlar_plus(inj_adi[i], ncomp, pa, fwhm, (good_values[i][2],good_values[i][3]), an_radius, plot=False)
    patches_plus.append(mlar)
```

```python
print(len(inj_adi))
```

Adding patches to data set with fake companions injected randomly within the previously found flux interval in ordre to reach the numbre of samples we want

```python
while len(patches_plus)<sample_nbr:
    adi_with_fake_comp, inj_posxy = inject_random_fake_comp(adi, psf, pxscale, pa, fwhm, an_radius, flx=flux_interv)
    snr_vale, f_source, sourcex, sourcey = evaluate_snr(adi_with_fake_comp, pa, fwhm, inj_posxy, an_radius)
    if snr_vale>snr_interv[0] and snr_vale<snr_interv[1]:
        mlar=make_mlar_plus(adi_with_fake_comp, ncomp, pa, fwhm, inj_posxy, an_radius, plot=False)
        patches_plus.append(mlar)
```

## C- class

```python
patches_minus=make_mlar_minus(adi, ncomp, pa, fwhm, an_radius, sample_nbr)
```

## Data augmentation

```python
patches_minus_aug=[]
patches_plus_aug=[]
```

Rotation

```python
for i in patches_minus:
    patches_minus_aug.append(patch_rotation(i))
    
for i in patches_plus:
    patches_plus_aug.append(patch_rotation(i))
```

Shift

```python
for i in patches_minus:
    patches_minus_aug.append(patch_shift(i))
    
for i in patches_plus:
    patches_plus_aug.append(patch_shift(i))
```

Average

```python
for i in range(len(patches_minus)):
    patches_minus_aug.append(patch_average(patches_minus))

for i in range(len(patches_plus)):
    patches_plus_aug.append(patch_average(patches_plus))
```

```python
np.save('patches_plus_'+str(an_nbr)+'L_D', patches_plus)
np.save('patches_plus_aug_'+str(an_nbr)+'L_D', patches_plus_aug)
np.save('patches_minus_'+str(an_nbr)+'L_D', patches_minus)
np.save('patches_minus_aug_'+str(an_nbr)+'L_D', patches_minus_aug)
```
